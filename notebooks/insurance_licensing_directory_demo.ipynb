# U.S. insurance licensing resource lookup

This runnable notebook demonstrates how to query TSI National's public 50-state licensing-resource dataset without treating the data as legal advice, regulator endorsement, or a guarantee that a course is required or approved.

- [Canonical 50-state licensing and CE directory](https://tsinational.com/insurance-licensing-and-ce-requirements-by-state/)
- [Public machine-readable CSV](https://milesdeepresearch.github.io/tsi-national-insurance-licensing-data/data/states.csv)
- [Dataset repository](https://github.com/milesdeepresearch/tsi-national-insurance-licensing-data)

Requirements and course availability can change. Use the returned primary-source links and confirm current instructions with the applicable regulator before applying.


```python
import pandas as pd

DATA_URL = 'https://milesdeepresearch.github.io/tsi-national-insurance-licensing-data/data/states.csv'
states = pd.read_csv(DATA_URL, dtype=str).fillna('')

required = {'state_code', 'state_name', 'license_guide_url', 'sellable', 'public_course_url'}
missing = required.difference(states.columns)
assert not missing, f'Missing required columns: {sorted(missing)}'
assert len(states) == 50 and states['state_code'].nunique() == 50
assert (states['sellable'].str.lower() == 'yes').sum() == 44
states[['state_code', 'state_name', 'sellable']].head()
```

```python
def lookup_state(code):
    code = str(code).strip().upper()
    match = states.loc[states['state_code'].str.upper() == code]
    if match.empty:
        raise ValueError(f'Unknown state code: {code}')
    row = match.iloc[0]
    return {
        'state': row['state_name'],
        'licensing_guide': row['license_guide_url'],
        'course_currently_listed': row['sellable'].lower() == 'yes',
        'current_course_route': row['public_course_url'] or None,
    }

lookup_state('TX')
```

## Reuse boundary

The notebook reads the current published dataset at runtime. It does not certify approval, eligibility, exam outcomes, or continued product availability. Preserve the source URLs and date your own downstream analysis.
